#### Figure 22: ‘Red Dead Redemption II’ Updates Impact on Active User Base Over Time

```python
plt.figure(figsize=(12, 6))

df_combined = red_dead_active_users.merge(red_dead_updates, left_on='Date', right_on='Date_e', how='outer').sort_values('Date')
df_combined['Number'] = df_combined['Number'].interpolate()

sns.lineplot(x='Date', y='Number', data=red_dead_active_users)

for index, row in red_dead_updates.iterrows():
    plt.axvline(x=row['Date_e'], color='r', linestyle='--', alpha=0.5)
    
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(custom_formatter))

from matplotlib.lines import Line2D

custom_lines = [Line2D([0], [0], color='red', lw=2, linestyle='--')]
plt.legend(custom_lines, ['Game Updates'], bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=1)

plt.title("Red Dead Redemption II updates impact on Users")
plt.xlabel('Years')
plt.ylabel('Users')

plt.tight_layout()
plt.show()
```

### Ongoing Support

#### Figure 23: The Impact of Updates on ‘Fortnite’ Active User Base

```python
event_dates = fortnite_updates1['event_date'].tolist()

plt.figure(figsize=(12, 6))

df_combined = fortnite_users1.merge(fortnite_updates1, left_on='date', right_on='event_date', how='outer').sort_values('date')
df_combined['users'] = df_combined['users'].interpolate()

sns.lineplot(x='date', y='users', data=df_combined)

for index, row in fortnite_updates1.iterrows():
    plt.axvline(x=row['event_date'], color='r', linestyle='--', alpha=0.5)
    
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(custom_formatter))
ax.set_xticks(event_dates)

from matplotlib.lines import Line2D

custom_lines = [Line2D([0], [0], color='red', lw=2, linestyle='--')]
plt.legend(custom_lines, ['Game Updates'], bbox_to_anchor=(0.5, -0.35), loc='upper center', ncol=1)

plt.xticks(rotation=90)
plt.title("Fortnite's Updates Impact on Users")


plt.tight_layout()
plt.show()
```

#### Figure 24: The Impact of Updates on ‘GTA V’ Active User Base
